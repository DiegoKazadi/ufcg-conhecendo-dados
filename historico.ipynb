```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Garantir que o estilo do Seaborn seja configurado
sns.set(style="whitegrid")

try:
    # Caminho para o arquivo tabela_historico_tratada.csv
    caminho_tabela_historico_tratada = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2/Dados/Tabelas_Tratadas/tabela_historico_tratada.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_tabela_historico_tratada):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela tabela_historico_tratada.csv
        historico_df = pd.read_csv(caminho_tabela_historico_tratada, low_memory=False)
        print("Tabela tabela_historico_tratada.csv carregada com sucesso!")

        # Visualizar as últimas 10 linhas
        print("\nÚltimas 10 linhas da tabela tabela_historico_tratada.csv:")
        print(historico_df.tail(10))

        # Identificar valores ausentes por coluna
        total_linhas = len(historico_df)
        valores_ausentes = historico_df.isnull().sum()
        percentuais_ausentes = (valores_ausentes / total_linhas) * 100
        valores_ausentes = valores_ausentes[valores_ausentes > 0]  # Filtrar apenas colunas com valores ausentes
        percentuais_ausentes = percentuais_ausentes[percentuais_ausentes > 0]  # Filtrar apenas colunas com valores ausentes

        if valores_ausentes.empty:
            print("Nenhum valor ausente encontrado na tabela.")
        else:
            print("\nValores ausentes por coluna (números absolutos):")
            print(valores_ausentes)
            
            print("\nPercentuais de valores ausentes por coluna:")
            print(percentuais_ausentes)

            # Criar um gráfico para visualizar os valores ausentes em percentual
            plt.figure(figsize=(12, 6))
            sns.barplot(
                x=percentuais_ausentes.index,
                y=percentuais_ausentes.values,
                palette="viridis"
            )
            plt.title("Percentual de Valores Ausentes por Coluna - Tabela Histórico", fontsize=16)
            plt.xlabel("Colunas", fontsize=12)
            plt.ylabel("Percentual de Valores Ausentes (%)", fontsize=12)
            plt.xticks(rotation=90)
            plt.tight_layout()
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.show()
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_tabela_historico_tratada}")

except Exception as e:
    print(f"Erro ao carregar, tratar ou visualizar valores ausentes: {e}")
```
