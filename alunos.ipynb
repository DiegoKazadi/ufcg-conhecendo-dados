# **Conhecer tabela alunos**

```python
import pandas as pd

# Caminho da tabela
caminho_tabela_alunos = r"E:\Mestrado UFCG\Semestre 2024.2\Dados\Tabelas_0\alunos.csv"

# Carregar a tabela
alunos = pd.read_csv(caminho_tabela_alunos, sep=';', encoding='utf-8')  # Ajuste o separador e encoding conforme necessário

# Exibir as primeiras linhas
display(alunos.head())

# Visualizar as colunas
print("Colunas da tabela 'alunos':")
for col in alunos.columns:
    print(f"- {col}")
```

###  Informações gerais da tabela

```python
# Ver tamanho da base
print(f"Número de alunos: {alunos.shape[0]}")
print(f"Número de variáveis (colunas): {alunos.shape[1]}")

# Visualizar tipos das variáveis
alunos.info()

# Primeiras linhas
alunos.head()
```

### variáveis de perfil dos alunos

### Gênero

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(data=alunos, x='GENERO', palette='pastel')
plt.title('Distribuição por Gênero')
plt.xticks(rotation=45)
plt.show()
```

### Estado Civil

```python
plt.figure(figsize=(8,5))
sns.countplot(data=alunos, x='ESTADO_CIVIL_ALUNOS', palette='coolwarm')
plt.title('Distribuição por Estado Civil')
plt.xticks(rotation=45)
plt.show()
```

### Nacionalidade

```python
print(alunos['NACIONALIDADE'].value_counts())
```

### Pais dos Estrangeiros

### Estado de origem

```python
plt.figure(figsize=(10,6))
sns.countplot(data=alunos, x='ESTADO', order=alunos['ESTADO'].value_counts().index, palette='viridis')
plt.title('Distribuição dos Alunos por Estado de Origem')
plt.xticks(rotation=45)
plt.show()
```

### Escolaridade prévia

```python
plt.figure(figsize=(6,4))
sns.countplot(data=alunos, x='TIPO_ENSINO_MEDIO', palette='Set2')
plt.title('Tipo de Escola de Ensino Médio')
plt.xticks(rotation=45)
plt.show()
```

### Ano de Conclusão do Ensino Médio

```python
plt.figure(figsize=(10,6))
sns.histplot(alunos['ANO_FORMATURA_ENSINO_MEDIO'], bins=30, kde=True, color='coral')
plt.title('Ano de Conclusão do Ensino Médio')
plt.xlabel('Ano')
plt.ylabel('Frequência')
plt.show()
```

### Situação acadêmica

```python
# Remover espaços dos nomes das colunas
alunos.columns = alunos.columns.str.strip()
# Mostrar as colunas reais do DataFrame
print(alunos.columns.tolist())
```

```python
situacoes = {
    'Ativos': alunos['ALUNOS_ATIVOS'].sum(),
    'Ex-Alunos': alunos['EX_ALUNOS'].sum(),
    'Inativos': alunos['ALUNOS_INATIVOS'].sum()
}

print(situacoes)
```

```python
situacoes = {
    'Ativos': alunos['ALUNOS_ATIVOS'].sum(),
    'Ex-Alunos': alunos['EX_ALUNOS'].sum(),
    'Inativos': alunos['ALUNOS_INATIVOS'].sum()
}

plt.figure(figsize=(6,4))
sns.barplot(x=list(situacoes.keys()), y=list(situacoes.values()), palette='rocket')
plt.title('Situação Atual dos Alunos')
plt.ylabel('Número de Alunos')
plt.show()
```
