# *Conhecer tabela Matriculas*

```python
# Exemplo de carregamento, caso o arquivo seja CSV
import pandas as pd

# Caminho do arquivo de matrícula
caminho_matriculas = 'E:/Mestrado UFCG/Semestre 2024.2/Dados/Tabelas_0/matriculas.csv'

# Carregar o arquivo CSV
matriculas = pd.read_csv(caminho_matriculas, sep=";")
matriculas.head()  # Ver as primeiras linhas para verificar se carregou corretamente
```

```python
# Verificar as colunas da tabela de matrículas
matriculas.columns
```

### Distribuição de Matrículas por Status (Status da matrícula: aprovado, reprovado, etc.)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contar a quantidade de matrículas por status
status_count = matriculas['ESTATUS'].value_counts()

# Plotando um gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=status_count.index, y=status_count.values, palette="viridis")
plt.title('Distribuição de Matrículas por Status')
plt.xlabel('Status da Matrícula')
plt.ylabel('Número de Matrículas')
plt.xticks(rotation=45)
plt.show()
```

### Média de Notas por Disciplina

```python
# Calcular a média de notas por disciplina
media_notas = matriculas.groupby('NOME')['NOTA'].mean().sort_values(ascending=False)

# Plotando as 10 disciplinas com maiores médias de nota
plt.figure(figsize=(10, 6))
media_notas.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Disciplinas com Maior Média de Nota')
plt.xlabel('Disciplina')
plt.ylabel('Média das Notas')
plt.xticks(rotation=45)
plt.show()
```

### Distribuição de Matrículas por Tipo de Matrícula

```python
# Contar a quantidade de matrículas por tipo
tipo_count = matriculas['TIPO'].value_counts()

# Plotando um gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=tipo_count.index, y=tipo_count.values, palette="Blues")
plt.title('Distribuição de Matrículas por Tipo')
plt.xlabel('Tipo de Matrícula')
plt.ylabel('Número de Matrículas')
plt.xticks(rotation=45)
plt.show()
```
