# **Carregar tabela dados pessoas**

```python
import pandas as pd
import os

try:
    # Caminho para o arquivo tabela_dados_pessoais_tratada.csv
    caminho_tabela_dados_pessoais_tratada = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2/Dados/Tabelas_Tratadas/tabela_dados_pessoais_tratada.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_tabela_dados_pessoais_tratada):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela tabela_dados_pessoais_tratada.csv
        tabela_dados_pessoais_tratada_df = pd.read_csv(caminho_tabela_dados_pessoais_tratada, low_memory=False)
        print("Tabela tabela_dados_pessoais_tratada.csv carregada com sucesso!")

        # Visualizar as últimas 10 linhas
        print("\nÚltimas 10 linhas da tabela tabela_dados_pessoais_tratada.csv:")
        print(tabela_dados_pessoais_tratada_df.tail(10))
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_tabela_dados_pessoais_tratada}")

except Exception as e:
    print(f"Erro ao carregar ou visualizar a tabela tabela_dados_pessoais_tratada.csv: {e}")
```

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Garantir que o estilo do Seaborn seja configurado
sns.set(style="whitegrid")

try:
    # Caminho para o arquivo tabela_dados_pessoais_tratada.csv
    caminho_tabela_dados_pessoais_tratada = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2/Dados/Tabelas_Tratadas/tabela_dados_pessoais_tratada.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_tabela_dados_pessoais_tratada):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela tabela_dados_pessoais_tratada.csv
        dados_pessoais_df = pd.read_csv(caminho_tabela_dados_pessoais_tratada, low_memory=False)
        print("Tabela tabela_dados_pessoais_tratada.csv carregada com sucesso!")

        # Visualizar as últimas 10 linhas
        print("\nÚltimas 10 linhas da tabela tabela_dados_pessoais_tratada.csv:")
        print(dados_pessoais_df.tail(10))

        # Identificar valores ausentes por coluna
        valores_ausentes = dados_pessoais_df.isnull().sum()
        valores_ausentes = valores_ausentes[valores_ausentes > 0]  # Filtrar apenas colunas com valores ausentes

        if valores_ausentes.empty:
            print("Nenhum valor ausente encontrado na tabela.")
        else:
            print("\nValores ausentes por coluna:")
            print(valores_ausentes)

            # Criar um gráfico para visualizar os valores ausentes
            plt.figure(figsize=(12, 6))
            sns.barplot(
                x=valores_ausentes.index,
                y=valores_ausentes.values,
                palette="viridis"
            )
            plt.title("Valores Ausentes por Coluna - Tabela Dados Pessoais", fontsize=16)
            plt.xlabel("Colunas", fontsize=12)
            plt.ylabel("Número de Valores Ausentes", fontsize=12)
            plt.xticks(rotation=90)
            plt.tight_layout()
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.show()
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_tabela_dados_pessoais_tratada}")

except Exception as e:
    print(f"Erro ao carregar, tratar ou visualizar valores ausentes: {e}")
```
