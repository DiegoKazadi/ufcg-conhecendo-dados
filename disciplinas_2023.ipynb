# **Carregar tabela disciplinas_2023**

```python
import pandas as pd

# Caminho para o arquivo de disciplinas_2023
caminho = r'E:\Mestrado UFCG\Semestre 2024.2\Dados\Tabelas_0\disciplinas_2023.csv'

# Carregar a tabela com o delimitador correto
disciplinas_2023 = pd.read_csv(caminho, encoding='ISO-8859-1', delimiter=';')

# Visualizar as primeiras linhas da tabela
print(disciplinas_2023.head())

# Visualizar as colunas da tabela
print(disciplinas_2023.columns)
```

###  Verificar a distribuição do número de créditos das disciplinas:

```python
import matplotlib.pyplot as plt

# Plotando a distribuição de créditos das disciplinas
plt.figure(figsize=(10, 6))
disciplinas_2023['CREDITO_DISCIPLINA'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribuição do Número de Créditos das Disciplinas', fontsize=14)
plt.xlabel('Número de Créditos', fontsize=12)
plt.ylabel('Quantidade de Disciplinas', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

### Verificar a quantidade de disciplinas por tipo (obrigatória, opcional, etc.)

```python
# Contagem das disciplinas por tipo
plt.figure(figsize=(8, 5))
disciplinas_2023['TIPO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Quantidade de Disciplinas por Tipo', fontsize=14)
plt.xlabel('Tipo de Disciplina', fontsize=12)
plt.ylabel('Quantidade de Disciplinas', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

###  Verificar a quantidade de disciplinas por semestre ideal:

```python
# Contagem de disciplinas por semestre ideal
plt.figure(figsize=(10, 6))
disciplinas_2023['SEMESTRE_IDEAL'].value_counts().sort_index().plot(kind='bar', color='salmon')
plt.title('Quantidade de Disciplinas por Semestre Ideal', fontsize=14)
plt.xlabel('Semestre Ideal', fontsize=12)
plt.ylabel('Quantidade de Disciplinas', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

### Análise das disciplinas obrigatórias vs opcionais:

```python
# Verificar a quantidade de disciplinas obrigatórias e opcionais
disciplinas_tipo = disciplinas_2023.groupby('TIPO')['CODIGO_DISCIPLINA'].count()
print(disciplinas_tipo)

# Plotar as disciplinas obrigatórias vs opcionais
plt.figure(figsize=(8, 5))
disciplinas_tipo.plot(kind='bar', color=['#FF6347', '#4682B4'])
plt.title('Disciplinas Obrigatórias vs Opcionais', fontsize=14)
plt.xlabel('Tipo de Disciplina', fontsize=12)
plt.ylabel('Quantidade de Disciplinas', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```
