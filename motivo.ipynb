```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Garantir que o estilo do Seaborn seja configurado
sns.set(style="whitegrid")

try:
    # Caminho para o arquivo tabela_motivo_evasao_tratada.csv
    caminho_tabela_motivo_evasao_tratada = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2/Dados/Tabelas_Tratadas/tabela_motivo_evasao_tratada.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_tabela_motivo_evasao_tratada):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela tabela_motivo_evasao_tratada.csv
        motivo_evasao_df = pd.read_csv(caminho_tabela_motivo_evasao_tratada, low_memory=False)
        print("Tabela tabela_motivo_evasao_tratada.csv carregada com sucesso!")

        # Visualizar as primeiras 5 linhas
        print("\nPrimeiras 5 linhas da tabela tabela_motivo_evasao_tratada.csv:")
        print(motivo_evasao_df.head())

        # Verificar se a coluna RAZAO_DE_INATIVIDADE_DE_ALUNO está presente
        if "RAZAO_DE_INATIVIDADE_DE_ALUNO" in motivo_evasao_df.columns:
            # Contar a frequência de cada motivo de inatividade
            frequencia_razoes = motivo_evasao_df["RAZAO_DE_INATIVIDADE_DE_ALUNO"].value_counts()

            # Visualizar os motivos e suas frequências
            print("\nFrequência dos motivos de inatividade:")
            print(frequencia_razoes)

            # Criar um gráfico de barras para a frequência dos motivos de inatividade
            plt.figure(figsize=(12, 6))
            sns.barplot(
                x=frequencia_razoes.index,
                y=frequencia_razoes.values,
                palette="coolwarm"
            )
            plt.title("Frequência dos Motivos de Inatividade", fontsize=16)
            plt.xlabel("Razões de Inatividade", fontsize=12)
            plt.ylabel("Número dos Alunos", fontsize=12)
            plt.xticks(rotation=45, ha="right")
            plt.tight_layout()
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.show()
        else:
            print("A coluna 'RAZAO_DE_INATIVIDADE_DE_ALUNO' não foi encontrada na tabela.")

    else:
        print(f"Arquivo não encontrado no caminho: {caminho_tabela_motivo_evasao_tratada}")

except Exception as e:
    print(f"Erro ao carregar, tratar ou visualizar os motivos de inatividade: {e}")
```

# Interpretação do Gráfico: Frequência dos Motivos de Inatividade

## Descrição
O gráfico de barras apresenta a distribuição da frequência dos motivos que levaram à inatividade de alunos em um determinado curso ou programa. Cada barra representa um motivo específico, e a altura indica a quantidade de alunos afetados por esse motivo.

## Resumo dos Motivos de Inatividade
1. **Desconhecido**:
   - Representa o maior motivo de inatividade, com aproximadamente 1.750 ocorrências.
   - A alta frequência sugere problemas de registro ou categorização dos dados.

2. **Abandono**:
   - Segundo maior motivo, com cerca de 500 a 600 ocorrências.
   - Indica um número significativo de alunos que deixaram o curso sem completá-lo.

3. **Expulsão**:
   - Frequência semelhante ao abandono, com valores próximos de 500 ocorrências.
   - Destaca a necessidade de analisar as razões para essas expulsões.

4. **Reingresso**:
   - Motivo com menor impacto, com cerca de 200 a 300 ocorrências.
   - Refere-se a alunos que saíram e depois retornaram ao curso.

5. **Desistência**:
   - Representa uma menor proporção, com menos de 200 ocorrências.
   - Pode estar associada a dificuldades pessoais ou acadêmicas.

6. **Concluiu, mas não colou grau**:
   - Frequência muito baixa, indicando menos de 100 casos.
   - Sinaliza situações em que o aluno completou os requisitos, mas não formalizou a graduação.

7. **Transferência**:
   - Razoavelmente raro, com aproximadamente 60 ocorrências.
   - Refere-se a alunos que migraram para outros cursos ou instituições.

8. **Cancelamento**:
   - Menos de 10 ocorrências, indicando um motivo raro de inatividade.

9. **Outros**:
   - Representa os motivos menos frequentes, com apenas 2 casos registrados.

## Insights e Recomendações
- **Alta Frequência de "Desconhecido"**:
  - Sugere uma lacuna significativa nos registros. É fundamental revisar os dados e melhorar a categorização dos motivos de inatividade.
- **Abandono e Expulsão**:
  - Os altos índices indicam a necessidade de políticas de retenção mais eficazes, como suporte acadêmico e psicológico.
- **Baixa Frequência de Transferência e Cancelamento**:
  - Esses motivos são raros e não parecem ser um problema significativo para o curso.

## Conclusão
A análise do gráfico evidencia que a maior parte dos casos de inatividade é atribuída a motivos desconhecidos, abandono ou expulsão. A melhoria na qualidade dos dados e a implementação de estratégias de apoio aos alunos podem reduzir esses números e melhorar a retenção no curso.


```python
import pandas as pd
import os

# Caminho para o arquivo tabela_alunos_tratada.csv
caminho_alunos_tratada = os.path.expanduser(
    '~/Área de Trabalho/Semestre 2024.2/Dados/Tabelas_Tratadas/tabela_alunos_tratada.csv'
)

# Verificar se o arquivo existe
if os.path.exists(caminho_alunos_tratada):
    try:
        # Carregar a tabela
        alunos_df = pd.read_csv(caminho_alunos_tratada)
        
        # Verificar se as colunas necessárias existem
        colunas_necessarias = [
            'MATRICULA', 'ID_CIDADAO', 'CODE_14102100',
            'TERMO_ADMISSAO_ESTUDANTES', 'TERMO_ESTADO_ALUNO', 'ESTADO_ALUNO'
        ]
        if all(coluna in alunos_df.columns for coluna in colunas_necessarias):
            # Filtrar alunos cujo início do curso foi 2011.1
            alunos_2011_1 = alunos_df[alunos_df['TERMO_ADMISSAO_ESTUDANTES'] == '2011.1']
            
            # Filtrar alunos que estavam INATIVO em 2013.1 ou 2013.2
            alunos_inativos_2013 = alunos_2011_1[
                (alunos_2011_1['TERMO_ESTADO_ALUNO'].isin(['2013.1', '2013.2'])) &
                (alunos_2011_1['ESTADO_ALUNO'] == 'INATIVO')
            ]
            
            # Exibir os resultados
            if not alunos_inativos_2013.empty:
                print("Alunos que iniciaram em 2011.1 e estavam INATIVOS em 2013:")
                print(alunos_inativos_2013[['MATRICULA', 'ID_CIDADAO']])
                
                # Criar diretório de saída, se não existir
                caminho_saida_dir = os.path.expanduser('~/Área de Trabalho/Semestre 2024.2/Resultados')
                os.makedirs(caminho_saida_dir, exist_ok=True)
                
                # Salvar o resultado em um arquivo CSV
                caminho_saida = os.path.join(caminho_saida_dir, 'alunos_inativos_2013.csv')
                alunos_inativos_2013.to_csv(caminho_saida, index=False)
                print(f"\nResultados salvos em: {caminho_saida}")
            else:
                print("Nenhum aluno encontrado que atenda aos critérios.")
        else:
            print("As colunas necessárias não foram encontradas na tabela.")
    except Exception as e:
        print(f"Ocorreu um erro ao processar os dados: {e}")
else:
    print("O arquivo tabela_alunos_tratada.csv não foi encontrado.")
```
