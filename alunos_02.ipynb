### Identificar problemas em cada tabela

```python

# Carregar tabelas tratadas
caminho_tabelas = "~/Área de Trabalho/Semestre 2024.2_LSD/Dados/Tabelas_Tratadas"
tabelas = {}

# Identificar arquivos CSV no diretório
for arquivo in os.listdir(os.path.expanduser(caminho_tabelas)):
    if arquivo.endswith(".csv"):
        caminho_arquivo = os.path.join(os.path.expanduser(caminho_tabelas), arquivo)
        tabelas[arquivo] = pd.read_csv(caminho_arquivo)

# Verificar tipos de dados e valores ausentes para cada tabela
for tabela, df in tabelas.items():
    print(f"Tabela: {tabela}")
    print("Informações sobre os tipos de dados:")
    print(df.info())  # Tipos de dados
    print("\nDados ausentes por coluna:")
    print(df.isnull().sum())  # Valores ausentes
    print("-" * 80)
```

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Garantir que o estilo do Seaborn seja configurado
sns.set(style="whitegrid")

try:
    # Caminho para o arquivo tabela_alunos_tratada.csv
    caminho_alunos = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2_LSD/Dados/Tabelas_Tratadas/alunos_02.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_alunos):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela tabela_alunos_tratada.csv
        alunos_df = pd.read_csv(caminho_alunos, low_memory=False)
        print("Tabela alunos_02.csv carregada com sucesso!")

        # Identificar valores ausentes por coluna
        valores_ausentes = alunos_df.isnull().sum()
        valores_ausentes = valores_ausentes[valores_ausentes > 0]  # Filtrar apenas colunas com valores ausentes

        if valores_ausentes.empty:
            print("Nenhum valor ausente encontrado na tabela.")
        else:
            print("\nValores ausentes por coluna:")
            print(valores_ausentes)

            # Criar um gráfico para visualizar os valores ausentes
            plt.figure(figsize=(12, 6))
            sns.barplot(
                x=valores_ausentes.index,
                y=valores_ausentes.values,
                palette="viridis"
            )
            plt.title("Valores Ausentes por Coluna", fontsize=16)
            plt.xlabel("Colunas", fontsize=12)
            plt.ylabel("Número de Valores Ausentes", fontsize=12)
            plt.xticks(rotation=90)
            plt.tight_layout()
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.show()
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_alunos}")

except Exception as e:
    print(f"Erro ao carregar, tratar ou visualizar valores ausentes: {e}")
```
