# **Descricao de cada Tabelas**

```python
import pandas as pd
import os

try:
    # Caminho para o diretório onde estão os arquivos CSV
    diretorio = os.path.expanduser('~/Área de Trabalho/Semestre 2024.2_LSD/Dados/Tabelas_Tratadas')

    # Verificar os arquivos no diretório
    arquivos = [f for f in os.listdir(diretorio) if f.endswith('.csv')]
    print(f"Tabelas encontradas: {arquivos}")

    # Carregar e exibir informações de cada tabela
    for arquivo in arquivos:
        caminho_arquivo = os.path.join(diretorio, arquivo)
        try:
            # Carregar a tabela
            tabela = pd.read_csv(caminho_arquivo, low_memory=False)
            print(f"\nTabela: {arquivo}")
            print(f"Linhas: {tabela.shape[0]}, Colunas: {tabela.shape[1]}")
            print("\nTipos de dados das colunas:")
            print(tabela.dtypes)
            print("=" * 80)
        except Exception as e:
            print(f"Erro ao carregar a tabela {arquivo}: {e}")

except Exception as e:
    print(f"Erro ao acessar o diretório ou processar as tabelas: {e}")
```
