```python
import pandas as pd
import os

try:
    # Caminho para o arquivo alunos.csv
    caminho_alunos = os.path.expanduser('/media/diego/PIETRO HD/Mestrado UFCG/Semestre 2024.2/LSD/Tabelas/alunos.csv')

    # Verificar se o arquivo existe
    if os.path.exists(caminho_alunos):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela alunos.csv especificando o delimitador como ponto e vírgula
        alunos_df = pd.read_csv(caminho_alunos, sep=';', low_memory=False)
        print("Tabela alunos.csv carregada com sucesso!")

        # Visualizar as últimas 10 linhas
        print("\nÚltimas 10 linhas da tabela alunos.csv:")
        print(alunos_df.tail(10))
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_alunos}")

except Exception as e:
    print(f"Erro ao carregar ou visualizar a tabela alunos.csv: {e}")
```

```python
# Configurar o estilo do Seaborn
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

try:
    # Caminho para o arquivo alunos_tratada.csv
    caminho_alunos = os.path.expanduser('/media/diego/PIETRO HD/Mestrado UFCG/Semestre 2024.2/LSD/Tabelas/alunos.csv')

    # Verificar se o arquivo existe
    if os.path.exists(caminho_alunos):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela alunos_tratada.csv
        alunos_df = pd.read_csv(caminho_alunos, sep=';', low_memory=False)
        print("Tabela alunos.csv carregada com sucesso!")

        # Identificar valores ausentes por coluna
        valores_ausentes = alunos_df.isnull().sum()
        valores_ausentes = valores_ausentes[valores_ausentes > 0]  # Filtrar apenas colunas com valores ausentes

        if valores_ausentes.empty:
            print("Nenhum valor ausente encontrado na tabela.")
        else:
            print("\nValores ausentes por coluna:")
            print(valores_ausentes)

            # Criar um gráfico para visualizar os valores ausentes
            plt.figure(figsize=(12, 6))
            sns.barplot(
                x=valores_ausentes.index,
                y=valores_ausentes.values,
                palette="viridis"
            )
            plt.title("Valores Ausentes por Coluna", fontsize=16)
            plt.xlabel("Colunas", fontsize=12)
            plt.ylabel("Número de Valores Ausentes", fontsize=12)
            plt.xticks(rotation=45, ha="right")  # Melhor ajuste de rótulos
            plt.tight_layout()
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.show()
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_alunos}")

except Exception as e:
    print(f"Erro ao carregar, tratar ou visualizar valores ausentes: {e}")
```
