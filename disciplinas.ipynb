## **Carregar tabela disciplinas_1979**


```python
import pandas as pd
import os

try:
    # Caminho para o arquivo disciplinas_1979_tratada.csv
    caminho_disciplinas_1979 = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2/Dados/Tabelas_Tratadas/disciplinas_1979_tratada.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_disciplinas_1979):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela disciplinas_1979_tratada.csv
        disciplinas_1979_df = pd.read_csv(caminho_disciplinas_1979, low_memory=False)
        print("Tabela disciplinas_1979_tratada.csv carregada com sucesso!")

        # Visualizar as últimas 10 linhas
        print("\nÚltimas 10 linhas da tabela disciplinas_1979_tratada.csv:")
        print(disciplinas_1979_df.tail(10))
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_disciplinas_1979}")

except Exception as e:
    print(f"Erro ao carregar ou visualizar a tabela disciplinas_1979_tratada.csv: {e}")
```

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Garantir que o estilo do Seaborn seja configurado
sns.set(style="whitegrid")

try:
    # Caminho para o arquivo tabela_disciplinas_tratada.csv
    caminho_disciplinas_1990 = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2/Dados/Tabelas_Tratadas/disciplinas_1990_tratada.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_disciplinas_1990):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela tabela_disciplinas_tratada.csv
        tabela_disciplinas_df = pd.read_csv(caminho_disciplinas_1990, low_memory=False)
        print("Tabela tabela_disciplinas_1990_tratada.csv carregada com sucesso!")

        # Identificar valores ausentes por coluna
        valores_ausentes = tabela_disciplinas_df.isnull().sum()
        valores_ausentes = valores_ausentes[valores_ausentes > 0]  # Filtrar apenas colunas com valores ausentes

        if valores_ausentes.empty:
            print("Nenhum valor ausente encontrado na tabela.")
        else:
            print("\nValores ausentes por coluna:")
            print(valores_ausentes)

            # Criar um gráfico para visualizar os valores ausentes
            plt.figure(figsize=(12, 6))
            sns.barplot(
                x=valores_ausentes.index,
                y=valores_ausentes.values,
                palette="viridis"
            )
            plt.title("Valores Ausentes por Coluna - Tabela de Disciplinas", fontsize=16)
            plt.xlabel("Colunas", fontsize=12)
            plt.ylabel("Número de Valores Ausentes", fontsize=12)
            plt.xticks(rotation=90)
            plt.tight_layout()
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.show()
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_disciplinas_1990}")

except Exception as e:
    print(f"Erro ao carregar, tratar ou visualizar valores ausentes: {e}")
```

## **Carregar tabela disciplinas_2017.csv**

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Garantir que o estilo do Seaborn seja configurado
sns.set(style="whitegrid")

try:
    # Caminho para o arquivo tabela_disciplinas_tratada.csv
    caminho_disciplinas = os.path.expanduser(
        '~/Área de Trabalho/Semestre 2024.2_LSD/Dados/Tabelas_Tratadas/disciplinas_2017.csv'
    )

    # Verificar se o arquivo existe
    if os.path.exists(caminho_disciplinas):
        print("O arquivo foi encontrado. Carregando agora...")
        
        # Carregar a tabela tabela_disciplinas_tratada.csv
        disciplinas_df = pd.read_csv(caminho_disciplinas, low_memory=False)
        print("Tabela disciplinas_2017.csv carregada com sucesso!")

        # Identificar valores ausentes por coluna
        valores_ausentes = disciplinas_df.isnull().sum()
        valores_ausentes = valores_ausentes[valores_ausentes > 0]  # Filtrar apenas colunas com valores ausentes

        if valores_ausentes.empty:
            print("Nenhum valor ausente encontrado na tabela.")
        else:
            print("\nValores ausentes por coluna:")
            print(valores_ausentes)

            # Criar um gráfico para visualizar os valores ausentes
            plt.figure(figsize=(12, 6))
            sns.barplot(
                x=valores_ausentes.index,
                y=valores_ausentes.values,
                palette="viridis"
            )
            plt.title("Valores Ausentes por Coluna - Tabela de Disciplinas", fontsize=16)
            plt.xlabel("Colunas", fontsize=12)
            plt.ylabel("Número de Valores Ausentes", fontsize=12)
            plt.xticks(rotation=90)
            plt.tight_layout()
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.show()
    else:
        print(f"Arquivo não encontrado no caminho: {caminho_disciplinas}")

except Exception as e:
    print(f"Erro ao carregar, tratar ou visualizar valores ausentes: {e}")
```
